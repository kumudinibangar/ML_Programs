```python
# AIM - Write a program to perform linear filtering on medical images
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading axial CT scan
# Check for noisy image
image = cv2.imread('/content/imgs.jpg', 0)
# 0 is taken as grayscale

print(image, image.dtype, image.shape)
# image array , data type and image shape

kernel = np.ones((5, 5)) / 25
# averaging filter (5 by 5)

convImage = cv2.filter2D(image, -1, kernel)
blur = cv2.blur(image, (3, 3))
gaussian = cv2.GaussianBlur(image, (37, 37), 0)
median = cv2.medianBlur(image, 5)


```

```python
()
# Display images
titles = ['Original Image', 'Convolved Image', 'Blur', 'Gaussian Blur', 'Median Blur']
outputs = [image, convImage, blur, gaussian, median]
plt.figure(figsize=(15, 10), dpi=80)

# for loop for image looping
for i in range(5):
 plt.subplot(2, 3, i + 1)
 plt.imshow(outputs[i], cmap='gray')
 plt.title(titles[i])

plt.tight_layout()
plt.show()
```
